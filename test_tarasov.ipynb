```
gaussian := Exp[ -(p-p0)^2/(2 sigma^2)];
tarasov = Integrate[Exp[-(x+p-p0)^2/(2 sigma^2)] Exp[-x/tau],{x,0,Infinity},Assumptions->{sigma>0,tau>0,p0>0}]
tarasov0 = tarasov/.{p->p0};
tarasovN = Simplify[tarasov/tarasov0]; (* Normalize at the peak position.) 
```

```
LogPlot[{gaussian,tarasovN}/.{p0->40,tau->1,sigma->1},{p,30,50}] (* Both function meet at p0 *)
```
